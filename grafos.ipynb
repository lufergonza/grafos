####**RECORRIDO DFS(ANCHURA)**

```python
# Define the decision tree as a dictionary
tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H', 'I'],
    'E': ['J', 'K'],
    'F': ['L', 'M'],
    'G': ['N', 'O'],
    'H': [], 'I': [], 'J': [], 'K': [],
    'L': [], 'M': [], 'N': [], 'O': []
}
```

```python
from collections import deque

def bfs(grafo, nodo, visitados=None):
    if visitados is None:
        visitados = set()

    cola = deque([nodo])
    visitados.add(nodo)

    while cola:
        actual = cola.popleft()
        print(actual)

        for vecino in grafo[actual]:
            if vecino not in visitados:
                visitados.add(vecino)
                cola.append(vecino)
```

```python

bfs(tree, 'A')
```

####**RECORRIDO POR DFS(PROFUNFIDAD)**

```python
def dfs(grafo, nodo, visitados=None):
    if visitados is None:
        visitados = set()

    visitados.add(nodo)
    print(nodo)

    for vecino in grafo[nodo]:
        if vecino not in visitados:
            dfs(grafo, vecino, visitados)


```

```python
dfs(tree,"A")
```
